## Wordcloud from article titles stored in as excel sheet coloum

1. import libraries

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image
import pandas as pd
```

2. Load exel dataset

```python
df = pd.read_excel("C:/Users/naihin/Personal/Publication/N_leaching_global/RAW.xlsx")
```

3. Read in your article titles data

```python
article_titles = df["title"]
```

4. Join the article titles into a single string

```python
text = " ".join(article_titles)
```

5. Create a word cloud object

```python
wordcloud = WordCloud().generate(text)
```

6. Plot and save the word cloud image file

```python
wordcloud = WordCloud(background_color="black", max_words=300).generate(text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
#plt.show()
plt.savefig("wordcloud_titles.tiff", dpi=300)
```
